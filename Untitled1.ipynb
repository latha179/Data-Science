```python
#split the data into Testing Data and Training Data
from sklearn.model_selection import train_test_split
```

```python
x=np.array(df[['Time','Amount']])
```

```python
y=np.array(df[['Class']])
```

```python
from sklearn.tree import DecisionTreeClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
#Model Training
model=DecisionTreeClassifier()  
```

```python
model.fit(x_train,y_train)
```

```python
#Model Evaluation
print('Accuracy score on Training data:',model.score(x_train,y_train))
```

```python
print('Accuracy score on Testing data:',model.score(x_test,y_test))
```

```python
data=np.array([[60.0,100.78]])
a=model.predict(data)
print(a)
if(a==0):
    print("No Fraud Transaction")
else:
    print("Fraud Transaction")
```
